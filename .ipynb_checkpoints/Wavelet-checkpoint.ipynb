### Project functions

```python
def evaluation(ground_truth: list, identified: list, window: int) -> (float, float):
    """
    The evaluation function takes the positions of ground truth labels, the positions of interest points
    identified by the model and the size of the error window within which deviations are tolerated.
    
    It will compute the precision and recall and return them in that order.
    """
    if len(identified) == 0:
        return 0, 0
    if len(ground_truth) == 0:
        return 1, 1
    recall_range = [i for x in identified for i in list(range(round(x)-window,round(x)+window))]
    acc_range = [i for x in ground_truth for i in list(range(round(x)-window,round(x)+window))]
    acc = 0
    recall = 0
    for x in ground_truth :
        if round(x) in recall_range:
            recall += 1
    for x in identified:
        if round(x) in acc_range:
            acc += 1
    acc /= len(identified)
    recall /= len(ground_truth)
    return acc, recall

def F1_measure(obtained: list, ground_truth: list) -> float:
    """
    Computes the F1 score from the model output and the ground truth labels.
    """
    pre, recall = evaluation(ground_truth=ground_truth, identified=obtained, window=20)
    if (pre + recall) == 0:
        output = 0
    else:
        output = (pre * recall) / (pre + recall)
    return output * 2

def generate_wavelet_space(signal: np.array, max_scale: int = 60) -> np.ndarray:
    """
    For a given signal, apply Continuous Wavelet Transformation with 2nd and 4th order Gaussian derivative
    wavelet functions up to certain scale, default to 60, and return the wavelet coefficient space in the 
    form of a 2D matrix.
    """
    coef2, _ = pywt.cwt(signal,np.arange(1,max_scale),'gaus2')
    coef4, _ = pywt.cwt(signal,np.arange(1,max_scale),'gaus4')
    output = np.maximum(coef2, coef4)
    return output


def maxima_detection(matrix: np.ndarray, neighbours: int = 1, gaussian_sigma: int = 2, pos_threshold: np.float32 = 0.05, visualization = True) -> [(int,int)]:
    """
    Given a 2D matrix of wavelet coefficient space, find all the local maxima, in the form of (time, scale).
    
    Variable 'neighbours' determines the side length from the center value that defines a local neighbourhood, default to 1 (so the local nighbourhood is 3*3).
    Variable 'gaussian_sigma' determines the sigma value used to apply Gaussian smoothing to the input matrix, default to 2.
    Variable 'pos_threshold' determines the threshold for all local maxima, default to 0.05, but should be empirically tested.
    Variable 'visualization' determines whether visualization of the distribution of wavelet coefficients will be produced.
    
    """
    assert type(matrix) == np.ndarray, "Input must be a numpy.ndarray!"
    assert len(matrix.shape) == 2, "Input matix must have two dimensions!"
    matrix = gaussian_filter(matrix, sigma=gaussian_sigma)
    matrix = np.pad(matrix, [(0, 2), (0, 0)], mode='constant')
    output = []
    coef = []
    row = matrix.shape[0]       
    col = matrix.shape[1]
    for i in range(neighbours, (row - neighbours)):
        a = list(range(i - neighbours, i + neighbours + 1))
        for j in range(neighbours, (col - neighbours)):
            b = list(range(j - neighbours, j + neighbours + 1))
            n = [(x,y) for x in a for y in b]
            n.remove((i,j))
            maxima = map(lambda x: matrix[i,j] >= matrix[x[0],x[1]], n)
            if all(maxima):
                coef.append(matrix[i,j])
                if (matrix[i,j] > pos_threshold):
                    output.append((i,j))
    for x in list(output):
        diff = [math.sqrt((y[1] - x[1]) ** 2 + (y[0] - x[0]) ** 2) for y in list(output)]
        diff = [(lambda x: (x <= 15) & (x != 0.0))(x) for x in diff]
        if any(diff):
            output.remove(x)
    if visualization:
        graph= sns.displot(coef, kde=True, bins=30)
        graph.refline(x=pos_threshold,
              color = "red",
              lw = 1.5)
        graph.set(xlabel='Wavelet Coefficient', ylabel='Count')
        plt.show()
    return set(output)

def construct_windows(peaks: [(int, int)]) -> [(int,(int, int))]:
    """
    Based on scale value, find the detection window for each interest points, output will be in the form (peak_time, (window_begin, window_end)).
    """
    output = []
    for scale, time in peaks:
        low = time - round(math.log(abs(scale))**3)
        high = time + round(math.log(abs(scale))**3)
        output.append((time,(low, high)))
    return output


def HOG_from_interval(data: np.ndarray, bins: int, lower_bound: int = 35) -> [int]:
    """
    For a given interval, compute the HOG representation of the given interval.
    Variable 'bins' determines the number of bins in the final output.
    Variable 'lower_bound' determines minimum length of the interval, any interval shorter will be padded to this length, default to 35.
    
    """
    if len(data) < lower_bound:
        data = signal.resample(data, int(lower_bound))
    histogram = np.zeros(bins)
    magnitude = np.gradient(data)
    angles = np.rad2deg(np.arctan(magnitude))
    magnitude = magnitude/np.linalg.norm(magnitude)
    assert len(angles) == len(magnitude)
    for idx, val in enumerate(angles):
        seg = int(divmod(val, int(180/bins))[0])
        if seg < 0:
            seg = bins - abs(seg)
        histogram[seg] += magnitude[idx]
    return histogram

def process_window_info(data: np.ndarray, windows: [(int, int)], interval_num: int = 3, bin_num: int = 8) -> [int]:
    """
    Given the input signal and the (peak_time, (window_begin, window_end)) pairs of interest point regions, return the HOG descriptor for all the regions.
    Variable 'data' is the input signal.
    Variable 'windows' is the (peak_time, (window_begin, window_end)) pairs of interest point regions.
    Variable 'interval_num' determines the number of overlapping intervals within each region.
    Variable 'bin_num' determines the number of bins used in the HOG of each intervals.
    
    """
    def divide_window_into_interval(window: (int, int), no_of_interval: int) -> int:
        # returns the size of each interval within the window.
        size = window[1] - window[0] + 1
        overlap = (no_of_interval - 1)/2
        num = no_of_interval - overlap
        return int(size/num)
    data = (data-min(data))/(max(data)-min(data))
    data = data * 10
    histograms = []
    x_points = []
    for peak, window in windows:
        if len(data) < window[1] or window[0] < 0:
            # print(len(data), window[1], window[0])
            # print("A window is removed because it's out of bounds!!")
            continue
        mean = np.mean(data[window[0]:window[1]])
        width = divide_window_into_interval(window, no_of_interval=interval_num)
        increment = int(width / 2) + 1
        # increment = int(width)
        HOG = []
        for x in range(interval_num):
            begin = window[0] + x * increment
            end = min(window[1], begin + width)
            if len(data[begin:end]) < 5:
                temp = list(HOG_from_interval(data[begin - 3:end + 3], bin_num))
            else:
                temp = list(HOG_from_interval(data[begin:end], bin_num))
            HOG += temp
        if len(HOG) > 0:
            if data[peak] > mean:
                x_points.append(peak)
                HOG = np.array(HOG)
                histograms.append(HOG)
            else:
                x_points.append(peak)
                HOG = np.array(HOG) * -1
                histograms.append(HOG)
    return histograms, x_points

def Cluster_Interest_Points(signal, peak_threshold, trough_threshold, cluster_num, max_scale = 60, visualization = True):
    coef = generate_wavelet_space(signal, max_scale=max_scale)
    peaks = maxima_detection(coef,pos_threshold=peak_threshold, visualization = visualization)
    if visualization:        
        fig = plt.figure(figsize=(12, 8))
        ax1 = plt.subplot(211)
        ax1.imshow(coef, aspect='auto')
        ax1.scatter(list(zip(*peaks))[1], list(zip(*peaks))[0], color = 'hotpink')
        ax2 = plt.subplot(212, sharex = ax1)
        ax2.plot(signal)
        for x in list(zip(*peaks))[1] :
            ax2.axvline(x = x, color = 'hotpink')
        plt.show()
    coef = generate_wavelet_space(-signal, max_scale=max_scale)
    troughs = maxima_detection(coef,pos_threshold=trough_threshold, visualization = visualization)
    if visualization:     
        fig = plt.figure(figsize=(12, 8))
        ax1 = plt.subplot(211)
        ax1.imshow(coef, aspect='auto')
        ax1.scatter(list(zip(*troughs))[1], list(zip(*troughs))[0], color = 'deepskyblue')
        ax2 = plt.subplot(212, sharex = ax1)
        ax2.plot(signal)
        for x in list(zip(*troughs))[1] :
            ax2.axvline(x = x, color = 'deepskyblue')
        plt.show()
        
    interest_points = list(peaks) + list(troughs)
    histograms, x_points = process_window_info(signal, construct_windows(interest_points), interval_num=3, bin_num=10)
    X = np.array(histograms)
    X = np.nan_to_num(X, nan=0, posinf=999, neginf=-999)
    model = KMeans(n_clusters=cluster_num, random_state=123, n_init='auto')
    kmeans = model.fit(X = X)
    output = []
    for x in range(cluster_num):
        cluster = np.where(kmeans.labels_ == x)[0]
        interest_points = [ x_points[i] for i in cluster ] 
        output.append(interest_points)
    return output
```

### Model validation with ECG data:

```python
# Read the data from files:
data = pd.read_csv(r'dataset/ludb.csv')
normal = data.loc[data['Conduction abnormalities'].isnull() & data['Extrasystolies'].isnull() & data['Cardiac pacing'].isnull()]
normal = normal.loc[normal['Rhythms'] == "Sinus rhythm"]
normal = list(normal["ID"])

abnormal = data.loc[data['Rhythms'] != "Sinus rhythm"]
abnormal = list(abnormal["ID"])

model_performance = pd.DataFrame(columns=['Label', 'Precision', 'Recall'])
```

### Normal ECG signals:
First we look at the 90 normal ECG signals:

```python
import warnings
warnings.filterwarnings("ignore")
detectors = Detectors(500)
for idx, patient_num in enumerate(normal):
    print("Processing patient number", idx, "out of", len(normal))
    suffix = ['avf', 'avl', 'avr', 'i', 'ii', 'iii', 'v1', 'v2', 'v3', 'v4', 'v5', 'v6']
    for channel in range(len(suffix)):
        stretch_factor = 1
        filename = r"dataset/ECG/"+str(patient_num)
        rec = wfdb.rdrecord(filename)
        ann = wfdb.rdann(filename, suffix[channel], shift_samps=True)
        # sig, fields = wfdb.rdsamp(filename, channels=[5])
        # xqrs = processing.XQRS(sig=sig[:,0], fs=fields['fs'])
        # xqrs.detect()
        # wfdb.plot_items(signal=sig, ann_samp=[xqrs.qrs_inds])
        annotation_begin = min(ann.sample)
        annotation_end = max(ann.sample)
        data = np.array(rec.p_signal.T[channel])
        R = ann.sample[[x == 'N' for x in ann.symbol]]
        T = ann.sample[[x == 't' for x in ann.symbol]]
        P = ann.sample[[x == 'p' for x in ann.symbol]]
        data = signal.resample(data, int(len(data) * stretch_factor))
        R = R * stretch_factor
        T = T * stretch_factor
        P = P * stretch_factor
        #============================================================
        # NeuroKit2
        _, rpeaks = nk.ecg_peaks(data, sampling_rate=500)
        Rpeaks = rpeaks['ECG_R_Peaks']
        Rpeaks = [x for x in Rpeaks if not (np.isnan(x) == True) and annotation_begin<x and annotation_end >x]
        try:
            _, waves_peak = nk.ecg_delineate(data, rpeaks, sampling_rate=500, method="peak")
            Tpeaks = waves_peak['ECG_T_Peaks']
            Tpeaks = [x for x in Tpeaks if not (np.isnan(x) == True) and annotation_begin<x and annotation_end >x]
            Ppeaks = waves_peak['ECG_P_Peaks']
            Ppeaks = [x for x in Ppeaks if not (np.isnan(x) == True) and annotation_begin<x and annotation_end >x]
        except:
            Tpeaks = []
            Ppeaks = []
            
        Rprecision, Rrecall = evaluation(ground_truth=R, identified=Rpeaks, window=20)
        Tprecision, Trecall = evaluation(ground_truth=T, identified=Tpeaks, window=20)
        Pprecision, Precall = evaluation(ground_truth=P, identified=Ppeaks, window=20)
        R_row = pd.Series({"Label": "NeuroKit2 - R peak", "Precision": Rprecision, "Recall": Rrecall})
        T_row = pd.Series({"Label": "NeuroKit2 - T peak", "Precision": Tprecision, "Recall": Trecall})
        P_row = pd.Series({"Label": "NeuroKit2 - P peak", "Precision": Pprecision, "Recall": Precall})
        model_performance = model_performance.append(R_row, ignore_index=True)
        model_performance = model_performance.append(T_row, ignore_index=True)
        model_performance = model_performance.append(P_row, ignore_index=True)        
        #============================================================
        # Other ECG detectors:
        try:
            hamilton = detectors.hamilton_detector(data)
            hamilton = [x for x in hamilton if annotation_begin<x and annotation_end >x]
        except:
            hamilton = []   
            
        try:
            christov = detectors.christov_detector(data)
            christov = [x for x in christov if annotation_begin<x and annotation_end >x]
        except:
            christov = []   
            
        try:
            engzee = detectors.engzee_detector(data)
            engzee = [x for x in engzee if annotation_begin<x and annotation_end >x]
        except:
            engzee = []   
            
        try:
            pan_tompkins = detectors.pan_tompkins_detector(data)
            pan_tompkins = [x for x in pan_tompkins if annotation_begin<x and annotation_end >x]
        except:
            pan_tompkins = []    
            
        try:
            swt = detectors.swt_detector(data)
            swt = [x for x in swt if annotation_begin<x and annotation_end >x]
        except:
            swt = []   
            
        try:
            average = detectors.two_average_detector(data)
            average = [x for x in average if annotation_begin<x and annotation_end >x]
        except:
            average = []   
            
        try:
            wqrs = detectors.wqrs_detector(data)
            wqrs = [x for x in wqrs if annotation_begin<x and annotation_end >x]
        except:
            wqrs = []    
        
        Rprecision, Rrecall = evaluation(ground_truth=R, identified=hamilton, window=20)
        R_row = pd.Series({"Label": "Hamilton - R peak", "Precision": Rprecision, "Recall": Rrecall})
        model_performance = model_performance.append(R_row, ignore_index=True)

        Rprecision, Rrecall = evaluation(ground_truth=R, identified=christov, window=20)
        R_row = pd.Series({"Label": "Christov - R peak", "Precision": Rprecision, "Recall": Rrecall})
        model_performance = model_performance.append(R_row, ignore_index=True)

        Rprecision, Rrecall = evaluation(ground_truth=R, identified=engzee, window=20)
        R_row = pd.Series({"Label": "Engzee - R peak", "Precision": Rprecision, "Recall": Rrecall})
        model_performance = model_performance.append(R_row, ignore_index=True)

        Rprecision, Rrecall = evaluation(ground_truth=R, identified=pan_tompkins, window=20)
        R_row = pd.Series({"Label": "Pan_Tompkins - R peak", "Precision": Rprecision, "Recall": Rrecall})
        model_performance = model_performance.append(R_row, ignore_index=True)

        Rprecision, Rrecall = evaluation(ground_truth=R, identified=swt, window=20)
        R_row = pd.Series({"Label": "SWT - R peak", "Precision": Rprecision, "Recall": Rrecall})
        model_performance = model_performance.append(R_row, ignore_index=True)

        Rprecision, Rrecall = evaluation(ground_truth=R, identified=average, window=20)
        R_row = pd.Series({"Label": "Two_average - R peak", "Precision": Rprecision, "Recall": Rrecall})
        model_performance = model_performance.append(R_row, ignore_index=True)

        Rprecision, Rrecall = evaluation(ground_truth=R, identified=wqrs, window=20)
        R_row = pd.Series({"Label": "WQRS - R peak", "Precision": Rprecision, "Recall": Rrecall})
        model_performance = model_performance.append(R_row, ignore_index=True)
        #============================================================
        # Proposed models:
        try:
            clusters_1 = Cluster_Interest_Points(data, 0.25, 0.5, 4, visualization=False)
        except:
            clusters_1 = []
        try:
            clusters_2 = Cluster_Interest_Points(data, 0.25, 0.5, 5, visualization=False)
        except:
            clusters_2 = []
        try:
            clusters_3 = Cluster_Interest_Points(data, 0.25, 0.5, 6, visualization=False)
        except:
            clusters_3 = []
        clusters = clusters_1 + clusters_2 + clusters_3 
        def remove_out_of_bound(cluster):
            cluster = np.array(cluster)
            output = cluster[(annotation_end > cluster) & (annotation_begin < cluster)]
            return output
        
        clusters = [remove_out_of_bound(x) for x in clusters]
        R_scores = [evaluation(ground_truth=R, identified=x, window=20) for x in clusters]
        Rprecision, Rrecall = R_scores[np.argmax(np.sum(R_scores, axis=1))]
        R_row = pd.Series({"Label": "Proposed model - R peak", "Precision": Rprecision, "Recall": Rrecall})
        model_performance = model_performance.append(R_row, ignore_index=True)
        
        
        T_scores = [evaluation(ground_truth=T, identified=x, window=20) for x in clusters]
        Tprecision, Trecall = T_scores[np.argmax(np.sum(T_scores, axis=1))]
        T_row = pd.Series({"Label": "Proposed model - T peak", "Precision": Tprecision, "Recall": Trecall})
        model_performance = model_performance.append(T_row, ignore_index=True)
        
        P_scores = [evaluation(ground_truth=P, identified=x, window=20) for x in clusters]
        Pprecision, Precall = P_scores[np.argmax(np.sum(P_scores, axis=1))]
        P_row = pd.Series({"Label": "Proposed model - P peak", "Precision": Pprecision, "Recall": Precall})
        model_performance = model_performance.append(P_row, ignore_index=True)
        
        # fig = plt.figure(figsize=(12, 8))
        # ax1 = plt.subplot(211)
        # ax1.plot(data)
        # for x in R_cluster:
        #     ax1.axvline(x = x, color = "hotpink")
        # for x in T_cluster:
        #     ax1.axvline(x = x, color = "deepskyblue")
        # for x in P_cluster:
        #     ax1.axvline(x = x, color = "lime")
        # plt.show()
        
```

```python
import warnings
warnings.filterwarnings("ignore")
detectors = Detectors(500)
model_performance = pd.DataFrame(columns=['Label', 'Precision', 'Recall'])
for idx, patient_num in enumerate(normal):
    print("Processing patient number", idx, "out of", len(normal))
    suffix = ['avf', 'avl', 'avr', 'i', 'ii', 'iii', 'v1', 'v2', 'v3', 'v4', 'v5', 'v6']
    for channel in range(len(suffix)):
        stretch_factor = 0.8
        filename = r"dataset/ECG/"+str(patient_num)
        rec = wfdb.rdrecord(filename)
        ann = wfdb.rdann(filename, suffix[channel], shift_samps=True)
        annotation_begin = min(ann.sample) * stretch_factor
        annotation_end = max(ann.sample) * stretch_factor
        data = np.array(rec.p_signal.T[channel])
        R = ann.sample[[x == 'N' for x in ann.symbol]]
        T = ann.sample[[x == 't' for x in ann.symbol]]
        P = ann.sample[[x == 'p' for x in ann.symbol]]
        data = signal.resample(data, int(len(data) * stretch_factor))
        R = R * stretch_factor
        T = T * stretch_factor
        P = P * stretch_factor        
        #============================================================
        # NeuroKit2
        _, rpeaks = nk.ecg_peaks(data, sampling_rate=500)
        Rpeaks = rpeaks['ECG_R_Peaks']
        Rpeaks = [x for x in Rpeaks if not (np.isnan(x) == True) and annotation_begin<x and annotation_end >x]
        try:
            _, waves_peak = nk.ecg_delineate(data, rpeaks, sampling_rate=500, method="peak")
            Tpeaks = waves_peak['ECG_T_Peaks']
            Tpeaks = [x for x in Tpeaks if not (np.isnan(x) == True) and annotation_begin<x and annotation_end >x]
            Ppeaks = waves_peak['ECG_P_Peaks']
            Ppeaks = [x for x in Ppeaks if not (np.isnan(x) == True) and annotation_begin<x and annotation_end >x]
        except:
            Tpeaks = []
            Ppeaks = []
            
        Rprecision, Rrecall = evaluation(ground_truth=R, identified=Rpeaks, window=20)
        Tprecision, Trecall = evaluation(ground_truth=T, identified=Tpeaks, window=20)
        Pprecision, Precall = evaluation(ground_truth=P, identified=Ppeaks, window=20)
        R_row = pd.Series({"Label": "NeuroKit2 - R peak", "Precision": Rprecision, "Recall": Rrecall})
        T_row = pd.Series({"Label": "NeuroKit2 - T peak", "Precision": Tprecision, "Recall": Trecall})
        P_row = pd.Series({"Label": "NeuroKit2 - P peak", "Precision": Pprecision, "Recall": Precall})
        model_performance = model_performance.append(R_row, ignore_index=True)
        model_performance = model_performance.append(T_row, ignore_index=True)
        model_performance = model_performance.append(P_row, ignore_index=True)        
        #============================================================
        # Other ECG detectors:
        try:
            hamilton = detectors.hamilton_detector(data)
            hamilton = [x for x in hamilton if annotation_begin<x and annotation_end >x]
        except:
            hamilton = []   
            
        try:
            christov = detectors.christov_detector(data)
            christov = [x for x in christov if annotation_begin<x and annotation_end >x]
        except:
            christov = []   
            
        try:
            engzee = detectors.engzee_detector(data)
            engzee = [x for x in engzee if annotation_begin<x and annotation_end >x]
        except:
            engzee = []   
            
        try:
            pan_tompkins = detectors.pan_tompkins_detector(data)
            pan_tompkins = [x for x in pan_tompkins if annotation_begin<x and annotation_end >x]
        except:
            pan_tompkins = []    
            
        try:
            swt = detectors.swt_detector(data)
            swt = [x for x in swt if annotation_begin<x and annotation_end >x]
        except:
            swt = []   
            
        try:
            average = detectors.two_average_detector(data)
            average = [x for x in average if annotation_begin<x and annotation_end >x]
        except:
            average = []   
            
        try:
            wqrs = detectors.wqrs_detector(data)
            wqrs = [x for x in wqrs if annotation_begin<x and annotation_end >x]
        except:
            wqrs = []    
        
        Rprecision, Rrecall = evaluation(ground_truth=R, identified=hamilton, window=20)
        R_row = pd.Series({"Label": "Hamilton - R peak", "Precision": Rprecision, "Recall": Rrecall})
        model_performance = model_performance.append(R_row, ignore_index=True)

        Rprecision, Rrecall = evaluation(ground_truth=R, identified=christov, window=20)
        R_row = pd.Series({"Label": "Christov - R peak", "Precision": Rprecision, "Recall": Rrecall})
        model_performance = model_performance.append(R_row, ignore_index=True)

        Rprecision, Rrecall = evaluation(ground_truth=R, identified=engzee, window=20)
        R_row = pd.Series({"Label": "Engzee - R peak", "Precision": Rprecision, "Recall": Rrecall})
        model_performance = model_performance.append(R_row, ignore_index=True)

        Rprecision, Rrecall = evaluation(ground_truth=R, identified=pan_tompkins, window=20)
        R_row = pd.Series({"Label": "Pan_Tompkins - R peak", "Precision": Rprecision, "Recall": Rrecall})
        model_performance = model_performance.append(R_row, ignore_index=True)

        Rprecision, Rrecall = evaluation(ground_truth=R, identified=swt, window=20)
        R_row = pd.Series({"Label": "SWT - R peak", "Precision": Rprecision, "Recall": Rrecall})
        model_performance = model_performance.append(R_row, ignore_index=True)

        Rprecision, Rrecall = evaluation(ground_truth=R, identified=average, window=20)
        R_row = pd.Series({"Label": "Two_average - R peak", "Precision": Rprecision, "Recall": Rrecall})
        model_performance = model_performance.append(R_row, ignore_index=True)

        Rprecision, Rrecall = evaluation(ground_truth=R, identified=wqrs, window=20)
        R_row = pd.Series({"Label": "WQRS - R peak", "Precision": Rprecision, "Recall": Rrecall})
        model_performance = model_performance.append(R_row, ignore_index=True)
        #============================================================
        # Proposed models:
        try:
            clusters_1 = Cluster_Interest_Points(data, 0.25, 0.5, 4, visualization=False)
        except:
            clusters_1 = []
        try:
            clusters_2 = Cluster_Interest_Points(data, 0.25, 0.5, 5, visualization=False)
        except:
            clusters_2 = []
        try:
            clusters_3 = Cluster_Interest_Points(data, 0.25, 0.5, 6, visualization=False)
        except:
            clusters_3 = []
        clusters = clusters_1 + clusters_2 + clusters_3 
        def remove_out_of_bound(cluster):
            cluster = np.array(cluster)
            output = cluster[(annotation_end > cluster) & (annotation_begin < cluster)]
            return output
        
        clusters = [remove_out_of_bound(x) for x in clusters]
        R_scores = [evaluation(ground_truth=R, identified=x, window=20) for x in clusters]
        Rprecision, Rrecall = R_scores[np.argmax(np.sum(R_scores, axis=1))]
        R_row = pd.Series({"Label": "Proposed model - R peak", "Precision": Rprecision, "Recall": Rrecall})
        model_performance = model_performance.append(R_row, ignore_index=True)
        
        
        T_scores = [evaluation(ground_truth=T, identified=x, window=20) for x in clusters]
        Tprecision, Trecall = T_scores[np.argmax(np.sum(T_scores, axis=1))]
        T_row = pd.Series({"Label": "Proposed model - T peak", "Precision": Tprecision, "Recall": Trecall})
        model_performance = model_performance.append(T_row, ignore_index=True)
        
        P_scores = [evaluation(ground_truth=P, identified=x, window=20) for x in clusters]
        Pprecision, Precall = P_scores[np.argmax(np.sum(P_scores, axis=1))]
        P_row = pd.Series({"Label": "Proposed model - P peak", "Precision": Pprecision, "Recall": Precall})
        model_performance = model_performance.append(P_row, ignore_index=True)
        # fig = plt.figure(figsize=(12, 8))
        # ax1 = plt.subplot(211)
        # ax1.plot(data)
        # for x in clusters[np.argmax(np.sum(R_scores, axis=1))]:
        #     ax1.axvline(x = x, color = "hotpink")
        # for x in R:
        #     ax1.axvline(x = x, color = "deepskyblue", alpha = 0.5)
        # # for x in clusters[np.argmax(np.sum(P_scores, axis=1))]:
        # #     ax1.axvline(x = x, color = "lime")
        # plt.show()
        
```
